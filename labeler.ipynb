# Data Labeling

```python
import pandas as pd
import os
```

```python
DATA_FILE = "data/cleanHistory.csv"
LABEL_FILE = "data/labelHistory.csv"
```

```python
CATEGORY = {
    '1': 'Learning',
    '2': 'Entertainment',
    '3': 'Music',
    '4': 'Finance',
    '5': 'News',
    '6': 'Others'
}
```

```python
df = pd.read_csv(DATA_FILE)
df.head()
```

```python
if os.path.exists(LABEL_FILE):
    labelDf = pd.read_csv(LABEL_FILE)
    existing_url = set(labelDf['video_url'])
    remainingDf = df[~df['video_url'].isin(existing_url)]
else:
    labelDf = pd.DataFrame(columns=['title', 'video_url', 'label', 'label_desc'])
    remainingDf = df
len(remainingDf)
```

```python
for k, v in CATEGORY.items():
    print(f"{k}: {v}")
```

```python
newLabel = []
```

```python
sampleDf = remainingDf.sample(frac=1).reset_index(drop=True)
len(sampleDf)
```

```python
for index, row in sampleDf.iterrows():
    print("Title: ", row['title'])
    print("Channel: ", row['channel'])
    print()

    choice = input(f"{CATEGORY} CATEGORY: ").strip().lower()

    if choice == 'q':
        break
    else:
        newEntry = {
            'title': row['title'],
            'channel': row['channel'],
            'video_url': row['video_url'],
            'label': int(choice),
            'label_desc': CATEGORY[choice]
        }
        newLabel.append(newEntry)
```

```python
if newLabel:
    newDf = pd.DataFrame(newLabel)
    finalDf = pd.concat([labelDf, newDf], ignore_index=True)
    finalDf.to_csv(LABEL_FILE, index=False)
```
