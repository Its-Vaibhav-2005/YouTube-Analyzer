# Migrate

- `Class collapsing`

```python
import pandas as pd 
```

```python
LABELED_FILE = 'data/labelHistory.csv'
```

```python
LABEL_MAP = {
    1: 1, # Learning -> Learning
    2: 2, # Ent -> Entertainment (Notice the swap!)
    3: 3, # Music -> Music (Notice the swap!)
    4: 4, # Finance -> Others
    5: 4, # News -> Others
    6: 4  # Others -> Others
}
```

```python
NEW_DESC = {
    1: 'Learning',
    2: 'Entertainment',
    3: 'Music',
    4: 'Others'
}
```

```python
df = pd.read_csv(LABELED_FILE)
len(df)
```

```python
df['label'] = df['label'].map(LABEL_MAP)
df['label_desc'] = df['label'].map(NEW_DESC)
```

```python
df.head()
```

```python
df.to_csv(LABELED_FILE, index=False)
```
